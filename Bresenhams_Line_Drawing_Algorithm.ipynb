```python
import matplotlib.pyplot as plt
```

```python
def draw_line(x1, y1, x2, y2):
    
    sx = 1 if x1 < x2 else -1
    sy = 1 if y1 < y2 else -1
    
    is_steep = abs(y2 - y1) > abs(x2 - x1)
    
    if is_steep:
        x1, y1 = y1, x1
        x2, y2 = y2, x2
    
    dx = abs(x2 - x1)
    dy = abs(y2 - y1)
    
    p = 2 * dy - dx
    
    x_values = []
    y_values = []
    x = x1
    y = y1
    
    while True:
        if is_steep:
            x_values.append(y)
            y_values.append(x)
        else:
            x_values.append(x)
            y_values.append(y)
        
        if x == x2 and y == y2:
            break
            
        if p < 0:
            p += 2 * dy
        else:
            y += sy
            p += 2 * dy - 2 * dx
        
        x += sx
        
    return x_values, y_values
```

```python
def plot_func(x_values, y_values):
    plt.plot(x_values, y_values, color='blue', marker='o')
    plt.grid('true')
    plt.title("Bresenham's line drawing algorthm")
    plt.xlabel("x-axis")
    plt.ylabel("y-axis")
```

```python
x10,y10 = 1, 1
x20,y20 = 8, 4
x_values, y_values = draw_line(x10, y10, x20, y20)
```
