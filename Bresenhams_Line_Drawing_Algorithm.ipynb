```python
import matplotlib.pyplot as plt
```

```python
x1,y1 = 1,1
x2,y2 = 8,4
```

```python
is_steep = abs(y2-y1) > abs(x2-x1)
if is_steep:
    x1,y1 = y1,x1
    x2,y2 = y2,x2
```

```python
dx = abs(x2 - x1)
dy = abs(y2 - y1)
```

```python
p = 2*dy - dx
p
```

```python
x,y = x1,y1
x_values = []
y_values = []
while x!=x2 and y!=y2:
    if is_steep:
        x_values.append(y)
        y_values.append(x)
    else:
        x_values.append(x)
        y_values.append(y)
        
    if p < 0:
        p += 2*dy
    else:
        y += 1
        p += 2*dy - 2*dx
    x += 1
```

```python
x_values
```

```python
y_values
```

```python
plt.plot(x_values,y_values,color='blue',marker='o')
plt.grid('true')
plt.title("Bresenham's line drawing algorthm")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
```
