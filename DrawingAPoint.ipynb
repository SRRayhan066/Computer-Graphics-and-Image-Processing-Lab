```python
import matplotlib.pyplot as plt
import math
```

```python
x = input()
y = input()

x = math.floor(float(x))
y = math.floor(float(y))
```

```python
plt.plot(x,y,'o',color='blue')
plt.title('Single point printing')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.xlim(0,5)
plt.ylim(0,5)
plt.grid('true')
```

```python
plt.show()
```
