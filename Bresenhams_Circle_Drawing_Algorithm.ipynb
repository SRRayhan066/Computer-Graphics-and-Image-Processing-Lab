```python
import matplotlib.pyplot as plt
```

```python
def in_quardents(x_values, y_values):
    
    # for fourth quardent
    x_values.extend(x_values[::-1])
    y_values.extend([-y for y in y_values[::-1]])
    
    # for other half
    x_values.extend([-x for x in x_values[::-1]])
    y_values.extend(y_values[::-1])
```

```python
def center_alignment(x_values, y_values, cx, cy):
    x_modified_values = [x + cx for x in x_values]
    y_modified_values = [y + cy for y in y_values]
    
    return x_modified_values, y_modified_values
```

```python
def circle_drawing(cx, cy, r):
    
    p = 3 - 2 * r
    x = 0
    y = r 
    
    x_values = []
    y_values = []
    
    while x <= y:
        x_values.append(x)
        y_values.append(y)
        
        print(f"p : {p}")
        
        x += 1
        
        if p < 0:
            p += 4 * x + 6
        else:
            y -= 1
            p += 4 * (x - y) + 10
    
    reversed_x_values = x_values[::-1]
    reversed_y_values = y_values[::-1]
    
    x_values.extend(reversed_y_values)
    y_values.extend(reversed_x_values)
    
    in_quardents(x_values, y_values)
    
    x_values, y_values = center_alignment(x_values, y_values, cx, cy)
    
    return x_values, y_values
```

```python
def plot_func(x_values, y_values, cx, cy):
    plt.plot(x_values, y_values, marker='o', color='blue')
    plt.plot(cx, cy, marker='o', color='red')
    plt.title("Bresenham's circle drawing algorithm")
    plt.xlabel("x-axis")
    plt.ylabel("y-axis")
    plt.grid("true")
```

```python
#center and radius of a circle
cx, cy = 0, 0
r = 8

x_values, y_values = circle_drawing(cx, cy, r)
```

```python
print(f"x values :\n {x_values}")
print(f"y values :\n {y_values}")
```

```python
plot_func(x_values, y_values, cx, cy)
```

```python
# For another coordinate
cx, cy = 0, 4
r = 8

x_values, y_values = circle_drawing(cx, cy, r)
```

```python
plot_func(x_values, y_values, cx, cy)
```
