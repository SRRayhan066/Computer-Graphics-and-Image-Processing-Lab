```python
import matplotlib.pyplot as plt
import math
```

```python
x1,y1 = 2,2
x2,y2 = 7,5
```

```python
m = (y1-y2)/(x1-x2)
b = y1 - m*x1
```

```python
print(f"Slope :: {m}")
print(f"Intercept :: {b}")
```

```python
x_values = []
y_values = []
y_round_values = []
```

```python
if x1>x2:
        x1,x2 = x2,x1
        y1,y2 = y2,y1
        
while x1 <= x2:
    y = m * x1 + b
    x_values.append(x1)
    y_values.append(y)
    y_round_values.append(round(y))
    x1 +=1
```

```python
y_values
```

```python
plt.plot(x_values,y_values,color='blue',marker='o')
plt.plot(x_values,y_round_values,color='red',marker='o')
plt.grid('true')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Direct Line Drawing')
```
