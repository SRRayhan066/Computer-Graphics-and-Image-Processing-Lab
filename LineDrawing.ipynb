```python
import matplotlib.pyplot as plt
import math
```

```python
print("Enter first coordinate")
x1 = float(input("x1 : "))
y1 = float(input("y1 : "))

print("Enter second coordinate")
x2 = float(input("x2 : "))
y2 = float(input("y2 : "))
```

```python
m = (y1-y2)/(x1-x2)
b = y1 - m*x1
```

```python
print(f"Slope :: {m}")
print(f"Intercept :: {b}")
```

```python
x_values = []
y_values = []
y_round_values = []
```

```python
if x1>x2:
        x1,x2 = x2,x1
        y1,y2 = y2,y1
        
while x1 <= x2:
    y = m * x1 + b
    x_values.append(x1)
    y_values.append(y)
    y_round_values.append(round(y))
    x1 +=1
```

```python
plt.plot(x_values,y_values,color='blue')
plt.plot(x_values,y_round_values,color='red')
plt.grid('true')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Direct Line Drawing')
```
