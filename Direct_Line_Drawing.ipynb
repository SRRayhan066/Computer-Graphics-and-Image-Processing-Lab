```python
import matplotlib.pyplot as plt
import math
```

```python
def DLA(x1, y1, x2, y2):
    
    m = (y2 - y1) / (x2 - x1)
    b = y1 - m * x1
    
    x_values = []
    y_values = []
    y_round_values = []
    
    if x1 > x2:
        x1, x2 = x2, x1
        y1, y2 = y2, y1
        
    x = x1
    y = y1
    
    while x <= x2:
        x_values.append(x)
        y_values.append(y)
        y_round_values.append(round(y))
        x += 1
        y = m * x + b
    
    return x_values, y_values, y_round_values
```

```python
def plot_func(x_values, y_values, y_round_values):
    plt.plot(x_values, y_values, color='blue', marker='o')
    plt.plot(x_values, y_round_values, color='red', marker='o')
    plt.xlabel('x-axis')
    plt.ylabel('y-axis')
    plt.grid('true')
    plt.title('Direct Line drawing algorthm')
```

```python
x1, y1 = 2, 2
x2, y2 = 7, 5
x_values, y_values, y_round_values = DLA(x1, y1, x2, y2)
print(x_values)
print(y_values)
plot_func(x_values, y_values, y_round_values)
```
