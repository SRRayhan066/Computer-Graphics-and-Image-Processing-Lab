## Coordinate transformation

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def plot_func(p, r, title):
    plt.plot(p[0], p[1], color = 'blue', marker = 'o')
    plt.plot(r[0], r[1], color = 'red', marker = 'o')
    plt.axhline(0, color='black', lw=1.5)  # x-axis
    plt.axvline(0, color='black', lw=1.5) # y-axis
    plt.title(title)
    plt.xlabel("x-axis")
    plt.ylabel("y-axis")
    plt.grid("true")
```

## Translation

```python
def translation(p):
    tx, ty = -1, -1
    t = np.array([[tx],
                  [ty]])
    r = p - t
    
    return r
```

```python
p = np.array([[0, 1, 5],
              [0, 1, 2]]) # orginal points

r = translation(p)
```

```python
print(r)
```

```python
plot_func(p, r, "2D-translation")
```

## Scaling

```python
def scaling(p):
    sx, sy = 2, 2
    
    s = np.array([[1 / sx, 0],
                  [0, 1 / sy]])
    r = np.dot(s, p)
    
    return r
```

```python
p = np.array([[2, 4],
              [6, 8]])

r = scaling(p)
```

```python
print(r)
```

```python
plot_func(p, r, "2D-Scaling")
```

## Rotation

```python
def rotation(p):
    
    degree = 90
    radian = np.radians(degree)
    
    rm = np.array([[np.cos(radian), np.sin(radian)],
                   [- np.sin(radian), np.cos(radian)]])
    
    r = np.dot(rm, p)
    
    return r
```

```python
p = np.array([[0, 4],
              [0, 4]])

r = rotation(p)
```

```python
print(r)
```

```python
plot_func(p, r, "2D-Rotation")
```
