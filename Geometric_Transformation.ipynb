```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def plot_func(p, r, title):
    plt.plot(p[0], p[1], color = 'blue', marker = 'o')
    plt.plot(r[0], r[1], color = 'red', marker = 'x')
    plt.title(title)
    plt.xlabel("x-axis")
    plt.ylabel("y-axis")
    plt.grid("true")
```

## Translation

```python
def translation(p):
    tx, ty = 5, 0
    t = np.array([[tx],
                  [ty]])
    r = p + t
    
    return r
```

```python
p = np.array([[2, 4],
              [6, 8]]) # orginal points

r = translation(p)
```

```python
print(r)
```

```python
plot_func(p, r, "2D-translation")
```

## Scaling

```python
def scaling(p):
    sx, sy = 2, 2
    
    s = np.array([[sx, 0],
                  [0, sy]])
    r = np.dot(s, p)
    
    return r
```

```python
p = np.array([[2, 4],
              [6, 8]])

r = scaling(p)
```

```python
print(r)
```

```python
plot_func(p, r, "2D-Scaling")
```

## Rotation

```python
def rotation(p):
    
    degree = 90
    radian = np.radians(degree)
    
    rm = np.array([[np.cos(radian), -np.sin(radian)],
                   [np.sin(radian), np.cos(radian)]])
    
    r = np.dot(rm, p)
    
    return r
```

```python
p = np.array([[0, 4],
              [0, 4]])

rt = rotation(p[:, 1])

r = np.array([[p[0, 0], rt[0]],
              [p[1, 0], rt[1]]])
```

```python
print(r)
```

```python
plot_func(p, r, "2D-Rotation")
```
