```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
def plot_func(p, r, title):
    plt.plot(p[0], p[1], color = 'blue', marker = 'o')
    plt.plot(r[0], r[1], color = 'red', marker = 'o')
    plt.axhline(0, color='black', lw=1.5)  # x-axis
    plt.axvline(0, color='black', lw=1.5) # y-axis
    plt.title(title)
    plt.xlabel("x-axis")
    plt.ylabel("y-axis")
    plt.grid("true")
```

## Translation

```python
def translation(p):
    tx, ty = 10, 0
    t = np.array([[tx],
                  [ty]])
    r = p + t
    
    return r
```

```python
p = np.array([[-1, -4, -6],
              [-3, -1, -4]]) # orginal points

r = translation(p)
```

```python
print(r)
```

```python
plot_func(p, r, "2D-translation")
```

## Scaling

```python
def scaling(p):
    sx, sy = 2, 2
    
    s = np.array([[sx, 0],
                  [0, sy]])
    r = np.dot(s, p)
    
    return r
```

```python
p = np.array([[2, 4],
              [6, 8]])

r = scaling(p)
```

```python
print(r)
```

```python
plot_func(p, r, "2D-Scaling")
```

## Rotation

```python
def rotation(p):
    
    degree = 45
    radian = np.radians(degree)
    
    rm = np.array([[np.cos(radian), - np.sin(radian)],
                   [np.sin(radian), np.cos(radian)]])
    
    r = np.dot(rm, p)
    
    return r
```

```python
p = np.array([[1, 2, 6],
              [1, 2, 3]])

r = rotation(p)
```

```python
print(r)
```

```python
plot_func(p, r, "2D-Rotation")
```

## Reflection

```python
def reflection(p, degree):
    
    radian = np.radians(degree)
    
    M = np.array([[np.cos(2 * radian), np.sin(2 * radian)],
                  [np.sin(2 * radian), -np.cos(2 * radian)]])
    
    r = np.dot(M, p)
    
    return r
```

```python
# For y = -x or x + y = 0
degree = 0
p = np.array([[9, 6, 4, 9],
              [-3, -1, -4, -3]])

r = reflection(p, degree)
```

```python
print(r)
```

```python
plot_func(p, r, "2D-Reflection")
```

## Shearing

```python
def shearing(p):
    shx = 2
    shy = 3
    
    shm = np.array([[1, shx],
                    [shy, 1]])
    
    r = np.dot(shm, p)

        
    return r
    
```

```python
# For the rectangle
p = np.array([[0, 1, 1, 0, 0],
              [0, 0, 1, 1, 0]])

r = shearing(p)
```

```python
print(r)
```

```python
plot_func(p, r, "2D-Shearing")
```
