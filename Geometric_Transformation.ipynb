```python
import matplotlib.pyplot as plt
import numpy as np
```

## Translation

```python
def translation(p):
    tx, ty = 3, 2
    t = np.array([[tx], [ty]])
    r = p + t
    
    return r
```

```python
p = np.array([[2, 6], [6, 8]]) # orginal point

r = translation(p)
```

```python
print(r)
```

```python
plt.plot(p[0, 0], p[1, 0], color = 'blue', marker = 'o')
plt.plot(r[0, 0], r[1, 0], color = 'red', marker = 'x')
plt.title("2D-translation")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid("true")
```
